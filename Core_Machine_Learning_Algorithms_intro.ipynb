##1.Linear Regression
Linear regression is one of the most basic forms of machine learning and is used to predict numeric values. 

```Task: build a linear model to predict the survival rate of passangers from the titanic dataset.```

*Helpful documentation: https://www.tensorflow.org/tutorials/estimator/linear*

#### How it Works


Linear regression follows a very simple concept. If data points are related linearly, we can generate a line of best fit for these points and use it to predict future values.


````python
# Example of a dataset with one feature (X) and one label (Y)


import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,2.5,3,4])
y = np.array([1,4,7,9,15])
plt.plot(x,y,'ro')
plt.axis([0,6,0,20]) # X axis: 0,6, Y axis: 0,20
plt.show()


````

in above plot:  Data has a linear coorespondence. When the X value increases, so does the Y. Because of this relation we can create a line of best fit for this dataset. In this example our line will only use one input variable, as we are working with two dimensions (1 feature and output label). In larger datasets with more features```(total dimensions (5D) = number of features (x=4) + output label(Y=1 )``` our line will have more features and inputs.

"Line of best fit refers to a line through a scatter plot of data points that best expresses the relationship between those points." (https://www.investopedia.com/terms/l/line-of-best-fit.asp)

Here's a refresher on the equation of a line in 2D.

$ y = mx + b $

```Parameter: m = slope between X and Y, b = intercept```

- slope = dy/dx = slope between X and Y find by differentiating y w.r.t x

- intercept: point where the best fit line start

#### Training model: at training time when X, Y are given, model will learn the best parameters (slope and intercept)

#### Test: at test time, we will use the best paramters value and given X to predict Y

````python
# now predict the best fit for the above dataset

plt.plot(x, y, 'ro')
plt.axis([0, 6, 0, 20])
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))
plt.show()


````

```best fit line: all points are as close as on the line and dataset are split eually to left and right side from center.```
